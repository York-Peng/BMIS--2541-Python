# Answer Exercise-3

```python
df_ml1 = pd.read_csv('../Data/AllDataset/ml1mdata2000.csv')
df_ml1.head()
```

```python
df_ml1['missing'] = 0
df_ml1.loc[df_ml1.isna().any(axis=1), 'missing']=1
df_all_count = df_ml1.groupby(['movie_id'])['movie_id'].agg('count').reset_index(name='counts')
print(df_all_count.head())
df_missing = df_ml1.groupby(['movie_id'])['missing'].agg('sum').reset_index(name='missingcount')
print(df_missing.head())
```

```python
df_missing_all_merge = pd.merge(df_all_count,df_missing)
df_missing_all_merge
```

```python
df_missing_all_merge['proportion'] = df_missing_all_merge['missingcount']/df_missing_all_merge['counts']
df_missing_all_merge
```

```python
df_rating = df_ml1.groupby('movie_id')['rating'].mean().reset_index()
df_rating['absolute_rank'] = df_rating['rating'].rank(ascending=False)
df_rating['percentile_rank'] = df_rating['rating'].rank(ascending=False, pct=True)
```

```python
df_rating.columns = ['Movie ID', 'Average_rating', 'Absolute rank', 'Percentile rank']
df_rating
```

# Exercise 7

```python
df_world = pd.read_csv('../Large Data/Large Data Sets/worlddata.csv')
```

```python
wd_Drathrate = df_world.loc[df_world['IndicatorName'] == "Death rate, crude (per 1,000 people)",["CountryCode","Year","Value"]]
wd_Drathrate.rename(columns={"Value": "Death_Rate"},inplace=True)

wd_RuralPop = df_world.loc[df_world['IndicatorName'] == "Rural population",["CountryCode","Year","Value"]]
wd_RuralPop.rename(columns={"Value": "Rural_population"},inplace=True)

wd_TotalStock = df_world.loc[df_world['IndicatorName'] == "International migrant stock, total",["CountryCode","Year","Value"]]
wd_TotalStock.rename(columns={"Value": "Total_stock"},inplace=True)
```

```python
df_merged = pd.merge(wd_Drathrate,wd_RuralPop, on=["CountryCode","Year"], how="outer")
df_merged = pd.merge(df_merged,wd_TotalStock, on=["CountryCode","Year"], how="outer")
df_merged
```
