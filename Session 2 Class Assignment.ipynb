```python
leaf_columns = ["leaves", "leafspots-halo", "leafspots-marg", "leafspot-size", "leaf-shread", 
                "leaf-malf", "leaf-mild"]
df_soybean['leaf_disease'] = df_soybean[leaf_columns].sum(axis=1)
df_soybean['leaf_disease']
```

```python
df_soybean['leaf_category'] = pd.cut(df_soybean['leaf_disease'], bins=[0, 2, 5, 7], 
                                     labels=['low', 'medium', 'high'])
df_soybean
```

```python
for column in df_adult.columns:
    if df_adult[column].isna().sum() > 0 and df_adult[column].dtype == 'object':
        df_adult[column].fillna(df_adult[column].mode()[0], inplace=True)
    elif df_adult[column].isna().sum() > 0 and df_adult[column].dtype != 'object':
        df_adult[column].fillna(df_adult[column].median()[0], inplace=True)
```
