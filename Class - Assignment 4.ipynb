# Exercise - 3

```python
world_data = pd.read_csv('../Large Data/Large Data Sets/worlddata.csv')
```

```python
wdiselect = world_data[(world_data['IndicatorName']=="CO2 emissions (kt)") \
                       & (world_data['Year']==2010)]
wdiselect.head()
```

```python
data_to_plot = wdiselect[['CountryCode','Value']]
data_to_plot.head()
```

```python
import folium
```

```python
url = 'https://raw.githubusercontent.com/python-visualization/folium/master/examples/data'
world_geo = f'{url}/world-countries.json'
```

```python
map2 = folium.Map(location=[40, 0], zoom_start=0.5)
```

```python
folium.Choropleth(geo_data=world_geo,
                  data=data_to_plot,
             name='choropleth',
             columns=['CountryCode', 'Value'],
             key_on='feature.id',
             fill_color='YlGnBu', fill_opacity=0.7, line_opacity=0.2,
             legend_name='CO2 emissions (kt)',
                 threshold_scale=[data_to_plot['Value'].min(), data_to_plot['Value'].quantile(0.35), data_to_plot['Value'].median(),
                                   data_to_plot['Value'].quantile(0.9), data_to_plot['Value'].max()]).add_to(map2)
folium.LayerControl().add_to(map2)
```
